# Data Visualization

```python
data.head(20)
```

```python
#Distribution of IMDb Ratings
plt.figure(figsize=(10, 8))
sns.histplot(data["IMDb Rating"], bins=10, kde=True)
plt.title("Distribution of IMDb Ratings")
plt.xlabel("IMDb Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# Movies Released Over Time (Year)
plt.figure(figsize=(10, 6))
sns.countplot(x="Year", data=data,hue="Year", palette="coolwarm")
plt.title("Movies Released Over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of MetaScores
plt.figure(figsize=(10, 8))
sns.histplot(data["MetaScore"], bins=10, kde=True, color='green')
plt.title("Distribution of MetaScores")
plt.xlabel("MetaScore")
plt.ylabel("Frequency")
plt.show()
```

```python
# Set the style for the plots
sns.set(style="whitegrid")
```

```python
correlation_matrix = data[['IMDb Rating', 'Year', 'MetaScore', 'Duration (minutes)']].corr()

#Heatmap of correlation between variables
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap of Correlation Between Variables")
plt.show()
```

```python
#Bar Plot of Movies Per Genre
plt.figure(figsize=(10, 8))
sns.countplot(data,x="Genre",hue="Genre", palette="muted")
plt.title("Movies per Genre")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot of IMDb Ratings by Certificates
plt.figure(figsize=(10, 8))
sns.boxplot(x="Certificates", y="IMDb Rating", data=data,hue="Certificates", palette="coolwarm")
plt.title("Boxplot of IMDb Ratings by Certificates")
plt.show()
```

```python
# Bar Plot of Top 5 Directors by Average IMDb Rating
top_5_directors_imdb = data.groupby("Director")["IMDb Rating"].mean().nlargest(5)
plt.figure(figsize=(10, 8))
sns.barplot(x=top_5_directors_imdb.index, y=top_5_directors_imdb.values, color='red')
plt.title("Top 5 Directors by Average IMDb Rating")
plt.ylabel("Average IMDb Rating")
plt.xlabel("Director")
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin Plot of IMDb Rating by Genre
plt.figure(figsize=(10, 8))
sns.violinplot(x="Genre", y="IMDb Rating", data=data,hue="Genre", palette="pastel")
plt.title("Violin Plot of IMDb Rating by Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
# Pair Plot for IMDb Rating, MetaScore, and Duration
sns.pairplot(data[["IMDb Rating", "MetaScore", "Duration (minutes)"]], diag_kind="kde")
plt.suptitle("Pair Plot for IMDb Rating, MetaScore, and Duration", y=1.02)
plt.show()
```

```python
# Scatter Plot of MetaScore vs IMDb Rating with Linear Regression Fit
plt.figure(figsize=(10, 8))
sns.regplot(x="MetaScore", y="IMDb Rating", data=data, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Scatter Plot of MetaScore vs IMDb Rating with Linear Regression Fit")
plt.show()
```

```python
# Bar Plot of Top 5 Longest Movies
top_5_longest = data.nlargest(5, "Duration (minutes)")[["Title", "Duration (minutes)"]]
plt.figure(figsize=(10, 8))
sns.barplot(x=top_5_longest["Title"], y=top_5_longest["Duration (minutes)"], color="blue")
plt.title("Top 5 Longest Movies by Duration")
plt.ylabel("Duration (minutes)")
plt.xlabel("Movie Title")
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the occurrences of each genre
genre_counts = data['Genre'].value_counts()

# Plot a pie chart
plt.figure(figsize=(10, 10))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"), explode=[0.05]*len(genre_counts))
plt.title('Distribution of Movies by Genre')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```

```python
# Count the occurrences of each Certificates
Certificates_counts = data['Certificates'].value_counts()

# Plot a pie chart
plt.figure(figsize=(15, 15))
plt.pie(Certificates_counts, labels=Certificates_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"), explode=[0.05]*len(genre_counts))
plt.title('Distribution of Movies by Genre')
plt.axis('equal')# Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()

```
